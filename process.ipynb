```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import albumentations as A
import os
```

```python
RAW_DATA_PATH = 'raw/'
data_files = os.listdir(RAW_DATA_PATH)
```

```python
images = np.array([Image.open(RAW_DATA_PATH + file) for file in data_files if file.endswith('.tif')], dtype=np.float32)
```

```python
#convert to a range of 0 to 1 for albumentations
images -= np.min(images)
images /= np.max(images)
```

```python
plt.imshow(images[0])
```

```python
images.shape, np.min(images), np.max(images)
```

```python
transform = A.Compose([
    A.CropNonEmptyMaskIfExists(width=256, height=256),
])

sampled_images = []

mask = np.ones(images.shape[1:3], dtype=np.uint8)

for image in images:
    for _ in range(50):
        transformed = transform(image=image, mask=mask)['image']

        if np.max(transformed) - np.min(transformed) > 0.1:
            sampled_images.append(transformed)
            print(len(sampled_images))

sampled_images = np.array(sampled_images)
```

```python
#show a bunch of images
fig, axs = plt.subplots(5, 5, figsize=(5, 5))
for i in range(5):
    for j in range(5):
        axs[i, j].imshow(sampled_images[i*5 + j])
        axs[i, j].axis('off')
```

```python
#save the images
np.save('dataset/terrain_images.npy', sampled_images)
```
