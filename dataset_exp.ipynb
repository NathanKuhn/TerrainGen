```python
import numpy as np
import matplotlib.pyplot as plt
```

Load the terrain dataset from the `dataset` folder.

```python
heightmaps = np.load("dataset/terrain_dataset.npy")
```

```python
plt.imshow(heightmaps[0])
plt.colorbar(label="Height (m)")
```

Statistics for the dataset:

```python
print(f"Count: {heightmaps.shape[0]}")
print(f"Range: [{heightmaps.min():4}, {heightmaps.max():4}]")
print(f"Mean:  {heightmaps.mean():4.2f}")
print(f"Std:   {heightmaps.std():4.2f}")
```

Show a sample of the dataset

```python
np.random.seed(0)
np.random.shuffle(heightmaps)

num = 8

fig, axs = plt.subplots(num, num, figsize=(num, num))

for i in range(num):
    for j in range(num):
        height_map = heightmaps[i * num + j]
        axs[i, j].imshow(height_map)
        axs[i, j].axis("off")
```
